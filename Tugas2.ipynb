```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# interval
x = np.linspace(0, 10, 500)

# deret fungsi sinusoidal
f = np.sin(x) + 0.5*np.sin(2*x) + (1/3)*np.sin(3*x) + (1/4)*np.sin(4*x) + (1/5)*np.sin(5*x)
```

# 1. Hitung integral dan diferensial dari fungsi 'f' untuk interval x

```python
# Menghitung turunan numerik f'(x) menggunakan np.gradient()
f_prime = np.gradient(f, x)

# Menghitung integral numerik F(x) menggunakan np.cumsum() (integral dari f)
F = np.cumsum(f) * (x[1] - x[0])  # Integral numerik dengan np.cumsu
```

### **Turunan \( f'(x) \)**

- **Output dari `f_prime`** adalah **array turunan** yang menunjukkan perubahan fungsi pada setiap titik \( x \).
- **Interpretasi Output**:
  - **Nilai positif**: Fungsi \( f(x) \) **menaik** pada titik tersebut.
  - **Nilai negatif**: Fungsi \( f(x) \) **menurun** pada titik tersebut.
  - **Nilai mendekati 0**: Titik tersebut adalah **puncak** atau **lembah** (fungsi berhenti berubah sementara).

Output turunan:

```python
print("Turunan f'(x) untuk seluruh interval x:")
print(f_prime)
```

### **Integral \( F(x) \)**

- **Output dari `F`** adalah **array integral** yang menunjukkan **total area** di bawah kurva \( f(x) \) dari \( x = 0 \) hingga setiap titik \( x \).
- **Interpretasi Output**:
  - **Integral yang lebih besar** menunjukkan **lebih banyak area** yang terhitung di bawah kurva.
  - **Integral yang lebih kecil atau negatif** menunjukkan **area di bawah sumbu \( x \)**.

Output integral:


```python
print("Integral F(x) untuk seluruh interval x:")
print(F)
```

---

## Kesimpulan

- **Turunan \( f'(x) \)** memberikan informasi tentang **kecepatan perubahan** fungsi di setiap titik \( x \).
- **Integral \( F(x) \)** memberikan informasi tentang **total area** di bawah fungsi \( f(x) \) dari \( x = 0 \) hingga titik \( x \).
- Output berupa **array** menunjukkan nilai-nilai turunan dan integral pada setiap titik \( x \), memberikan gambaran tentang **perilaku fungsi** pada interval yang diberikan.

# 2. hitung secara terpisah bagian integral di atas dan di bawah y=0


```python
from scipy.integrate import trapezoid

# Membagi fungsi f(x) menjadi dua bagian: positif dan negatif
f_positive = np.where(f >= 0, f, 0)  # Bagian positif (f(x) >= 0)
f_negative = np.where(f < 0, f, 0)   # Bagian negatif (f(x) < 0)

# Menghitung integral positif dan negatif menggunakan aturan Trapesium
integral_positive = trapezoid(f_positive, x)  # Integral area positif
integral_negative = trapezoid(f_negative, x)  # Integral area negatif
```

**Output 1: Integral Area Positif**

```python
print(f"Integral area positif: {integral_positive}")
```

**Interpretasi Output**:
- Hasil **integral area positif** sebesar **4.604** menunjukkan total luas **area di atas sumbu \( x \)** untuk fungsi \( f(x) \) pada interval \( x = 0 \) hingga \( x = 10 \). Nilai positif ini mengindikasikan bahwa fungsi lebih banyak berada di atas sumbu \( x \) dalam interval tersebut.

---

**Output 2: Integral Area Negatif**


```python
print(f"Integral area negatif: {integral_negative}")
```

**Interpretasi Output**:
- Hasil **integral area negatif** sebesar **-2.418** menunjukkan total luas **area di bawah sumbu \( x \)** untuk fungsi \( f(x) \). Nilai negatif ini mengindikasikan bahwa fungsi lebih banyak berada di bawah sumbu \( x \) dalam interval tersebut.

---

### **Kesimpulan**
- **Integral Area Positif** memberi kita informasi tentang **luas area** yang terhitung di atas sumbu \( x \).
- **Integral Area Negatif** memberi kita informasi tentang **luas area** yang terhitung di bawah sumbu \( x \).
- Dengan menggunakan **metode trapesium**, kita bisa menghitung kedua area tersebut secara numerik untuk mendapatkan pemahaman yang lebih baik tentang bagaimana fungsi \( f(x) \) berperilaku di seluruh interval.

# 3. plot semua fungsi: f, f', F (atas dan bawah dengan area warna berbeda)


```python
# Plot semua fungsi (f, f', F)
plt.figure(figsize=(10, 8))

# Plot fungsi f(x)
plt.subplot(3, 1, 1)
plt.plot(x, f, label='f(x)', color='blue')
plt.fill_between(x, f, where=(f > 0), color='skyblue', alpha=0.5, label='Area above x-axis')
plt.fill_between(x, f, where=(f < 0), color='salmon', alpha=0.5, label='Area below x-axis')
plt.title('Fungsi f(x)')
plt.legend()

# Plot turunan f'(x)
plt.subplot(3, 1, 2)
plt.plot(x, f_prime, label="f'(x)", color='green')
plt.title("Turunan f'(x)")
plt.legend()

# Plot integral F(x)
plt.subplot(3, 1, 3)
plt.plot(x, F, label='Integral F(x)', color='purple')
plt.fill_between(x, F, color='lightcoral', alpha=0.5, label='Area above x-axis')
plt.fill_between(x, F, color='lightgreen', alpha=0.5, label='Area below x-axis')
plt.title('Integral F(x)')
plt.legend()

plt.tight_layout()
plt.show()
```
