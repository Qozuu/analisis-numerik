````python
import numpy as np
import matplotlib.pyplot as plt

# Number of points
n = 200
NPM = 120

# Generate random x values between 0 and 4
np.random.seed(NPM)
x = np.random.uniform(0, 4*np.pi, n)

# Original function
y_true = 4 * np.sin(2*x + 0.2)

# Add Gaussian noise
noise = np.random.normal(0, 0.4, n)  # mean=0, std=0.5
y_noisy = y_true + noise

# Plot
plt.scatter(x, y_noisy, label="Noisy data", color="red", alpha=0.6)
plt.plot(np.sort(x), 4*np.sin(4*np.sort(x) + 0.2), label="True function", color="blue")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

# TUGAS: temukan kembali fungsi asli dengan interpolasi dari titik-titik data acak (y_noisy)
````

# **JAWABAN**

### **1. Import Library**
Pertama, import dua library utama :
- **`numpy`** : Digunakan untuk perhitungan numerik dan manipulasi array.
- **`matplotlib.pyplot`** : Digunakan untuk menggambar grafik.

```
import numpy as np
import matplotlib.pyplot as plt
```

### **2. Menyiapkan Data Acak**
Kedua, data acak `x` dihasilkan dalam rentang \\(0\\) hingga \\(4\pi\\). Fungsi **`np.random.uniform()`** digunakan untuk menghasilkan angka acak dalam rentang yang telah ditentukan. Fungsi **`np.random.seed(NPM)`** memastikan bahwa hasil angka acak yang dihasilkan akan konsisten setiap kali kode dijalankan, sehingga data acak yang dihasilkan dapat direplikasi.

```
np.random.seed(NPM)
x = np.random.uniform(0, 4*np.pi, n)
```

### **3. Membuat Fungsi Asli dan Penambahan Noise**
Ketiga, menunjukkan fungsi asli yang digunakan untuk menghasilkan data, yaitu \\(y=4sin\\)\\((2x+0.2\\)). Kemudian, **noise Gaussian** ditambahkan untuk menciptakan data noisy, yang mensimulasikan data yang lebih realistis

```
y_true = 4 * np.sin(2*x + 0.2)
noise = np.random.normal(0, 0.4, n)  # mean=0, std=0.4
y_noisy = y_true + noise
```

### **4. Interpolasi Polinomial**
Keempat, interpolasi dilakukan menggunakan polinomial dengan berbagai orde :



````python
# Interpolation - Linear (Order 1)
coeffs_linear = np.polyfit(x, y_noisy, 1)  # Linear fit
poly_linear = np.poly1d(coeffs_linear)

# Interpolation - Quadratic (Order 2)
coeffs_quadratic = np.polyfit(x, y_noisy, 2)  # Quadratic fit
poly_quadratic = np.poly1d(coeffs_quadratic)

# Interpolation - Cubic (Order 3)
coeffs_cubic = np.polyfit(x, y_noisy, 3)  # Cubic fit
poly_cubic = np.poly1d(coeffs_cubic)
````

* **Interpolasi Linear (Orde 1)** : `np.polyfit(x, y_noisy, 1)` digunakan untuk mendapatkan koefisien polinomial orde 1. Kemudian, `poly_linear = np.poly1d(coeffs_linear)` mengubah koefisien tersebut menjadi objek polinomial yang bisa digunakan untuk prediksi.

* **Interpolasi Kuadrat (Orde 2) dan Kubik (Orde 3)** dilakukan dengan cara yang sama, tetapi menggunakan orde 2 dan 3 pada `np.polyfit` untuk menangkap pola yang lebih kompleks dalam data noisy.

### **5. Visualisasi Hasil**
Kelima, grafik yang menunjukkan data acak dengan noise dan fungsi asli, serta hasil interpolasi dengan berbagai orde polinomial digambar.

````python
plt.figure(figsize=(10, 6))

# Plot noisy data
plt.scatter(x, y_noisy, label="Noisy data", color="red", alpha=0.6)

# Plot the true function
plt.plot(np.sort(x), 4 * np.sin(2 * np.sort(x) + 0.2), label="True function", color="blue")

# Plot interpolations
x_sorted = np.sort(x)
plt.plot(x_sorted, poly_linear(x_sorted), label="Linear Interpolation", color="green", linestyle='--')
plt.plot(x_sorted, poly_quadratic(x_sorted), label="Quadratic Interpolation", color="purple", linestyle='--')
plt.plot(x_sorted, poly_cubic(x_sorted), label="Cubic Interpolation", color="orange", linestyle='--')

# Labels and legend
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

# Show plot
plt.show()
````

* **Titik Merah** : Menampilkan data yang telah diberi noise menggunakan `plt.scatter`.

* **Garis Biru** : Menampilkan fungsi asli yang digunakan untuk menghasilkan data menggunakan `plt.plot`.

* **Interpolasi Linear, Kuadrat, dan Kubik** : Masing-masing diplot dengan warna berbeda (hijau, ungu, oranye) untuk membandingkan seberapa baik masing-masing polinomial mendekati data asli.

### **6. Persamaan Polinomial**
Keenam, setelah interpolasi dilakukan, persamaan polinomial untuk masing-masing orde dicetak. Ini memberikan gambaran matematis tentang hasil interpolasi.

````python
print("Linear Interpolation Equation: ", poly_linear)
print("Quadratic Interpolation Equation: ", poly_quadratic)
print("Cubic Interpolation Equation: ", poly_cubic)
````

Setelah interpolasi dilakukan, persamaan polinomial dari setiap orde dicetak menggunakan `print()` untuk memberikan gambaran matematis tentang hasil interpolasi.


### **7. Keterangan Tambahan** :
* **Orde 1 (Linear)** : Cocok untuk model yang paling sederhana dan sering digunakan untuk data yang tidak terlalu kompleks.

* **Orde 2 (Kuadrat)** : Menangkap pola lebih baik dengan kurva parabola, cocok untuk data yang sedikit lebih kompleks.

* **Orde 3 (Kubik)** : Menggambarkan kurva yang lebih fleksibel, yang berguna untuk menangkap fluktuasi lebih lanjut dari data noisy.

**Tidak menampilkan Orde 4 dan Orde 5 karena** :
* **Overfitting** : Menggunakan orde polinomial yang lebih tinggi (orde 4 dan 5) cenderung menghasilkan model yang terlalu cocok dengan data yang ada (termasuk noise), yang disebut overfitting. Model ini akan mempelajari fluktuasi yang tidak relevan dalam data dan menjadi kurang generalis.

* **Kompleksitas Berlebihan** : Semakin tinggi orde polinomial, semakin kompleks modelnya. Hal ini justru dapat membuat model menjadi tidak realistis dan tidak berguna dalam konteks tugas ini, karena kita hanya ingin menggambarkan pola asli dari data.
